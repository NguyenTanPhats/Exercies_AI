```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
```

```python
import pandas as pd
url='/content/passenger.csv'
data= pd.read_csv(url)
data.head(10)
```

```python
data.hist()
plt.show()
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
model = LinearRegression()
model.fit(X=pd.DataFrame(data['Miles(X)']), y=data['Dollars (Y)'])

X = pd.DataFrame(data['Miles(X)'])

intercept = model.intercept_
coef = model.coef_

print('Intercept: ', intercept)
print('Coeffient: ', coef)
print('Độ chính xác: ', model.score(X, y=data['Dollars (Y)']))

y_pred = model.predict(X=pd.DataFrame(data['Miles(X)']))
```

```python

data.plot(kind='scatter', x='Miles(X)', y='Dollars (Y)', figsize=(9, 9), color='red')
plt.plot(data['Miles(X)'], y_pred, color='blue')
```

```python
import numpy as np
import pandas as pd
model = LinearRegression()

X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2, data['Miles(X)']**3] ).T

model.fit(X, y=data['Dollars (Y)'])

print('Intercept: ', model.intercept_)
print('Coeffient: ', model.coef_)
print('Độ chính xác: ', model.score(X, y=data['Dollars (Y)']))

range = np.arange(0, 5500, 100)
X_range = pd.DataFrame([range, range**2, range**3]).T

y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='Miles(X)', y='Dollars (Y)', figsize=(9, 9), color='red')
plt.plot(range, y_pred, color='blue')
```

```python
import numpy as np
import pandas as pd
model = LinearRegression()

X = pd.DataFrame([data['Miles(X)'], data['Miles(X)']**2, data['Miles(X)']**3, data['Miles(X)']**4, data['Miles(X)']**5]).T

model.fit(X, y=data['Dollars (Y)'])

print('Intercept: ', model.intercept_)
print('Coeffient: ', model.coef_)
print('Độ chính xác: ', model.score(X, y=data['Dollars (Y)']))

range = np.arange(0, 5500, 100)
X_range = pd.DataFrame([range, range**2, range**3, range**4, range**5]).T

y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='Miles(X)', y='Dollars (Y)', figsize=(9, 9), color='red')
plt.plot(range, y_pred, color='blue')
```

LogisticRegression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
data_classification = pd.read_csv('/content/classification.csv')
data_classification
```

```python
X = data_classification.iloc[:, 1:2]
y = data_classification.iloc[:, -1]
```

```python
X = np.asarray(X)
y = np.asarray(y)
```

```python
model = LogisticRegression(solver='liblinear', random_state=1)
model.fit(X, y)
y_pred = model.predict(X)

print('Intercept: ', model.intercept_)
print('Coeffient: ', model.coef_)
print('Độ chính xác: ', model.score(X, y))
```

```python
np.set_printoptions(precision=2)
cm = confusion_matrix(y, y_pred)
print('Số lượng dự đoán giữa y dự đoán và y thật:\n', cm)
```
